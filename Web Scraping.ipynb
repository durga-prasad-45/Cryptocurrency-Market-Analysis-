```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
import time

# Optional: run in headless mode
chrome_options = Options()
#chrome_options.add_argument("--headless")
chrome_options.add_argument("--start-maximized")

driver = webdriver.Chrome(options=chrome_options)

url = "https://coinmarketcap.com/"
driver.get(url)

# wait for page to load fully
time.sleep(5)

# Find all coin name elements
coins = driver.find_elements(By.XPATH, "//p[contains(@class,'coin-item-name')]")

coin_names = []

for coin in coins[:100]:
    coin_names.append(coin.text)

# Print first 100
for i, name in enumerate(coin_names, start=1):
    print(i, name)

print("\nTotal Extracted:", len(coin_names))

driver.quit()
```

```python
import requests
import pandas as pd

# API endpoint used by CoinMarketCap website
api_url = "https://api.coinmarketcap.com/data-api/v3/cryptocurrency/listing"

params = {
    "start": 1,
    "limit": 100,          # Change this if you want more
    "sortBy": "market_cap",
    "sortType": "desc",
    "convert": "USD"
}

response = requests.get(api_url, params=params)
data = response.json()

coins = data["data"]["cryptoCurrencyList"]

# Extract required fields
coin_data = []
for coin in coins:
    coin_data.append({
        "Rank": coin["cmcRank"],
        "Name": coin["name"],
        "Symbol": coin["symbol"],
        "Price (USD)": coin["quotes"][0]["price"]
    })

# Convert to DataFrame
df = pd.DataFrame(coin_data)

# Save to CSV
#df.to_csv("coinmarketcap_top100.csv", index=False)

print("CSV file created successfully!")
print("Total Rows:", len(df))
```

```python
import requests
import pandas as pd

api_url = "https://api.coinmarketcap.com/data-api/v3/cryptocurrency/listing"

params = {
    "start": 1,
    "limit": 1000,      # Change to 500 or 1000 if needed
    "sortBy": "market_cap",
    "sortType": "desc",
    "convert": "USD"
}

response = requests.get(api_url, params=params)
data = response.json()

coins = data["data"]["cryptoCurrencyList"]

coin_data = []

for coin in coins:
    quote = coin["quotes"][0]

    coin_data.append({
        "Rank": coin["cmcRank"],
        "Name": coin["name"],
        "Symbol": coin["symbol"],
        "Price (USD)": quote["price"],
        "1h %": quote["percentChange1h"],
        "24h %": quote["percentChange24h"],
        "7d %": quote["percentChange7d"],
        "Market Cap": quote["marketCap"],
        "Volume (24h)": quote["volume24h"]
    })

df = pd.DataFrame(coin_data)

df.to_csv("coinmarketcap_full_data.csv", index=False)

print("CSV file created successfully!")
print("Total Rows:", len(df))
```
