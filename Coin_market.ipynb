## Outliers Treatment

```python
column = df["Price (USD)"]
Q1 = column.quantile(0.25)
Q3 = column.quantile(0.75)
IQR = Q3 - Q1
lower_limit = Q1 - 1.5 * IQR
upper_limit = Q3 + 1.5 * IQR
outliers = df[(column < lower_limit) | (column > upper_limit)]
```

## Data Visualization

## Univariate Analysis

#### Why choose Price (USD)?
Price shows the value distribution of cryptocurrencies and helps identify skewness 

```python
plt.figure(figsize=(8,5))
sns.histplot(df['Price (USD)'], bins=40)
plt.title("Price Distribution")
plt.show()
```

#### observation
- Very tall bars near zero are representing most cryptocurrencies are priced very low
- Very small bars representing a few cryptocurrencies have very high prices

### Why choose Market Cap?
Market capitalization represents the total value of a cryptocurrency, which indicates its market dominance

```python
fig = px.box(
    df,
    x="Market Cap",
    title="Market Capitalization Distribution"
)

fig.show()
```

#### Observation
- Minimum Value (Left Whisker) represents the lowest market capitalization in the dataset
- First Quartile (Q1) – The left edge of the box represents the 25% of cryptocurrencies have market caps below this value
- Median (Middle Line in the Box) - The vertical line inside the box represents the 50% of cryptocurrencies have market caps below the median
- Third Quartile (Q3) – The right edge of the box represents the 75% of cryptocurrencies have market caps below this value
- Maximum Value (Right Whisker) - The right whisker represents the largest market capitalization values in the dataset

### Why choose Volume (24h)?
Trading volume shows how actively coins are traded in the market

```python
fig = px.violin(
    df,
    y="Volume (24h)",
    box=True,
    title="Distribution of Trading Volume"
)

fig.show()
```

### Observation
- Median trading volume is around 2.5 million units
- Majority of coins have low trading volume
- A few coins have extremely high trading activity

```python
fig = px.scatter(
    df,
    x="Price (USD)",
    y="Volume (24h)",
    color="Market Cap",
    hover_name="Symbol",
    title="Price vs Trading Volume of Cryptocurrencies"
)

fig.show()
```

### Bivariate Analysis

### Why these columns?
- Symbol → cryptocurrency identifier

```python
top10 = df.sort_values("Market Cap", ascending=False).head(10)

fig = px.bar(
    top10,
    x="Symbol",
    y="Market Cap",
    title="Top 10 Cryptocurrencies by Market Capitalization"
)

fig.show()
```

### Observation
- The result is a bar chart where longer bars indicate higher market caps
- Top few cryptocurrencies contribute to a large portion of the market value
- The market shows strong dominance by leading coins

```python
top20 = df.sort_values("Volume (24h)", ascending=False).head(20)

fig = px.funnel(
    top20,
    x="Volume (24h)",
    y="Symbol",
    title="Top Cryptocurrencies by Trading Volume"
)

fig.show()
```

```python
corr = df.corr(numeric_only=True)

fig = px.imshow(
    corr,
    text_auto=True,
    title="Correlation Matrix",
    width=600,
    height=600
)

fig.show()
```

### Observation
- Market cap and trading volume often show positive correlation
- Price change variables usually show weak correlations, indicating market volatility

## Pivot Tables
